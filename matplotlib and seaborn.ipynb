```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("netflix.csv")
```

```python
data
```

```python
sns.countplot(data=data, x="type")
plt.title("Distribution of Movies and TV Shows on Netflix")
plt.show()
```

```python
data['release_year'] = data['release_year'].astype(int)
release_count = data['release_year'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
sns.lineplot(x=release_count.index, y=release_count.values)
plt.title("Number of Netflix Releases Over the Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(data=data, x="rating")
plt.title("Distribution of ratings in  Netflix")
plt.show()
```

```python
print(data.describe())
```

```python
sns.histplot(data['listed_in'], kde=True)

plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
sns.scatterplot(x='release_year', y='type', data=data)

plt.title('Scatter Plot')
plt.show()
```

```python
sns.boxplot(x='type', y='release_year', data=data)
plt.title('Box Plot')
plt.show()
```

```python
sns.lineplot(x='release_year', y='country', data=data)
plt.title('Line Plot')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a figure with multiple subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Create individual plots
sns.histplot(data['release'], kde=True, ax=axes[0, 0])
sns.scatterplot(x='release_year', y='type', data=data, ax=axes[0, 1])
sns.barplot(x='release_year', y='type', data=data, ax=axes[1, 0])
sns.boxplot(x='release_year', y='type', data=data, ax=axes[1, 1])

# Customize individual plots here

# Set a common title for the entire figure
fig.suptitle('Multiple Plots in One Figure')
plt.show()
```
